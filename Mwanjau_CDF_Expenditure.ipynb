```python
import numpy as np

def read_expenditure_data(fileName = "CDF_EXPENDITURE_ON_PROJECTS.csv"):
    'Function to load excel data'
    import pandas as pd
    import numpy as np
    
    #Read csv file with the pandas read_CSV method
    
    cdf_exp = pd.read_csv(fileName)
    
    
    return cdf_exp

cdf_exp = read_expenditure_data()
```

```python
#Check for null values summary
cdf_exp.isna().sum()
```

```python
cdf_exp.describe()
```

```python
#Drops columns
dp_cols = ['econ2','remarks']

for dpcols in dp_cols:
    cdf_exp.drop(dpcols,axis=1,inplace=True)

cdf_exp.dtypes
```

```python

#replace all other instance with Other in econ1 column
cdf_exp['econ1'] = cdf_exp['econ1'].str.replace('other','Other')

#Remove rows with missing values and accounting for missing values
cdf_exp.dropna(axis=0, inplace=True)
cdf_exp.isna().sum()

```

```python
cdf_exp.describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(cdf_exp['econ1'], cdf_exp['estimated_cost'],'ro')
```

From above Most of the budget was sourced as 'Capital Expenditure': which seems right incomparison to Recurrent Expenditure and Others

```python
plt.plot(cdf_exp['econ1'],cdf_exp['total_amount'],'bo')
```

From above we can see that between 2003 and 2010 recurrent expenditure surpassed Capital expenditure indicating some fraudilent issues
