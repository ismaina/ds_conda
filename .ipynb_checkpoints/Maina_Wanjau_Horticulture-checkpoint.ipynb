```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
amp = pd.read_csv("C:\\Users\\maina\\Documents\\Digital Academy\\Data Science\\DS_Crash_Course_labs\\Average_Monthly_Price_of_Food_Crops_Commodities_Scheduled_Horticultural_Crops_between_2012_to_2015.csv")
```

```python
amp.head(5)
```

```python
amp['Values_in_Ksh'] = amp['Values_in_Ksh'].str.replace('KES','')
amp['Values_in_Ksh'] = amp['Values_in_Ksh'].astype('float64')
```

```python
amp.head()
```

```python
amp = amp.reindex(['OBJECTID', 'Commodity_Type', 'Produce_Variety', 'Unit', 'Volume_in_Kgs', 'Values_in_Ksh' ,'Date'], axis=1)
amp.columns
```

```python
amp.describe()
amp["Date"] = amp["Date"].astype('datetime64')
```

```python
amp.describe()
```

```python
amp['Commodity_Type'].value_counts().plot(kind='bar')
```

```python
amp['Produce_Variety'].value_counts().plot(kind='bar')
```

```python
pv = amp.groupby(['Produce_Variety'])[['Volume_in_Kgs','Values_in_Ksh']].sum()
sns.distplot(pv['Values_in_Ksh'],bins=10)
```

```python
#Pie chart to show total number of Commodities export and their percentages.
counts = amp['Produce_Variety'].value_counts()
fig=plt.figure(figsize=(7,7))
ax=fig.gca()
def make_autopct(counts):
    def my_autopct(pct):
        total = sum(counts)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
counts.plot.pie(autopct=make_autopct(counts), startangle=0, ax=ax)
ax.set_title('Total count of Commodities Exported.', color='#006600')
```
