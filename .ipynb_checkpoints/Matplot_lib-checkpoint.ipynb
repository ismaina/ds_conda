```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,10,11)
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('y Label')
plt.title('TItle')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])   #plot a blank canvas

axes.plot(x,y)
```

```python
#
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.set_title('Bigger Plot')
axes2.set_title('Smaller Plot')
```
