```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#Data head
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Distribution plot
#sns.distplot(tips['total_bill'],bins=100)
sns.distplot(tips['total_bill'],kde=False,bins=10)
```

```python
#join two dist plots eg compare distribution of total bills vs tip size
#Kind controls the scater plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') #linear fit; regresion
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#takes care of all the numerical column values
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

# Catgegorical Plots


```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python

sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
#similar to Pandas.count
```

```python
#Showing the distribution of Categorical data
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
#compare day to bills in comparision to smokers
```

```python
tips.head()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#strip plot will draw a scater plot where one variable is categorical
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
#similar to stripplot but points are adjusted so that they dont overlap
sns.swarmplot(x='day',y='total_bill',data=tips)

#combination of strip and violin plots
```

```python
#similar to stripplot but points are adjusted so that they dont overlap
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')

#combination of strip and violin plots
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
