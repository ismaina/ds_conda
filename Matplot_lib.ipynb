```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,10,11)
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('y Label')
plt.title('TItle')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])   #plot a blank canvas

axes.plot(x,y)
```

```python
#
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
#axes1.set_title('Bigger Plot')
#axes2.set_title('Smaller Plot')
axes1.plot(x,y,'r')
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

# Figure size and DPI

```python
myfig =  plt.figure()

axx = myfig.add_axes([0,0,1,1])
axx.plot(x,x**2,label='X Squared')

axx.plot(x,x**3, label='X Cubed')
axx.legend(loc=0) #Specify the legend location
```

```python
fig= plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green',linewidth=3, alpha =0.5) #You can use RGB Hex Code , linewudth is the thickness of the line,
#alpha  is for transparency
```

```python
#line style

fig= plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green',linewidth=3, linestyle='--')
```

```python
#marker size

fig= plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple',lw=0.5, linestyle='-', marker = 'o', markersize=20,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='blue')
```

```python
#Set the min and max of the axis

fig= plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple',lw=2, linestyle='--')

ax.set_xlim([0,2])
ax.set_ylim([0,1])
```
